```python
# Symmetric Binary Tree

# Problem Description
# Given a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

# Problem Constraints
# 1 <= number of nodes <= 105

# Input Format
# First and only argument is the root node of the binary tree.

# Output Format
# Return 0 / 1 ( 0 for false, 1 for true ).

# Example Input
# Input 1:
#     1
#    / \
#   2   2
#  / \ / \
# 3  4 4  3
# Input 2:
#     1
#    / \
#   2   2
#    \   \
#    3    3

# Example Output
# Output 1:
#  1
# Output 2:
#  0


# Example Explanation
# Explanation 1:
#  The above binary tree is symmetric. 
# Explanation 2:
# The above binary tree is not symmetric.
```

```python
# Definition for a  binary tree node
# class TreeNode:
#	def __init__(self, x):
#		self.val = x
#		self.left = None
#		self.right = None

class Solution:
	# @param A : root node of tree
	# @return an integer
    def helper(self,root1,root2):
        if root1==None and root2==None:
            return 1
        if root1!=None and root2==None:
            return 0
        if root1==None and root2!=None:
            return 0
        return self.helper(root1.left,root2.right) and self.helper(root1.right,root2.left)
        
        
	def isSymmetric(self, A):
        x=self.helper(A.left,A.right)
        return x

```

```python
# Invert the Binary Tree
# Problem Description
# Given a binary tree A, invert the binary tree and return it.
# Inverting refers to making the left child the right child and vice versa.

# Problem Constraints
# 1 <= size of tree <= 100000

# Input Format
# First and only argument is the head of the tree A.

# Output Format
# Return the head of the inverted tree.

# Example Input
# Input 1:
#      1
#    /   \
#   2     3
# Input 2:
#      1
#    /   \
#   2     3
#  / \   / \
# 4   5 6   7


# Example Output
# Output 1:
#      1
#    /   \
#   3     2
# Output 2:
#      1
#    /   \
#   3     2
#  / \   / \
# 7   6 5   4

# Example Explanation
# Explanation 1:
# Tree has been inverted.
# Explanation 2:
# Tree has been inverted.
```

```python
# Definition for a  binary tree node
# class TreeNode:
#	def __init__(self, x):
#		self.val = x
#		self.left = None
#		self.right = None

class Solution:
	# @param A : root node of tree
	# @return the root node in the tree
	def invertTree(self, A):
        if A==None:
            return None
        x=self.invertTree(A.left)
        y=self.invertTree(A.right)
        A.left=y
        A.right=x
        return A
```

```python
# Definition for a  binary tree node
# class TreeNode:
#    def __init__(self, x):
#        self.val = x
#        self.left = None
#        self.right = None

class Solution:
    # @param A : root node of tree
    # @param B : root node of tree
    # @return an integer

    def check(self,a, b):
        if a == None and b == None:
            return 1

        if a == None or b == None:
            return 0

        if a.val == b.val:
            return 1

        return 0


    def recur(self,A, B):
        ans = 0
        if self.check(A.left, B.left) and self.check(A.right, B.right):
            temp = 0
            if A.left != None:
                temp = self.recur(A.left, B.left)
            if temp == -1:
                return -1

            ans += temp
            temp = 0
            if A.right != None:
                temp = self.recur(A.right, B.right)
            if temp == -1:
                return -1

            ans += temp

        elif self.check(A.left, B.right) and self.check(A.right, B.left):
            temp = 0
            ans += 1
            if A.left != None:
                temp = self.recur(A.left, B.right)

            if temp == -1:
                return -1

            ans += temp
            temp = 0
            if A.right != None:
                temp = self.recur(A.right, B.left)

            if temp == -1:
                return -1

            ans += temp

        else:
            return -1

        return ans


# class Solution:
    # @param A : root node of tree
    # @param B : root node of tree
    # @return an integer
    def solve(self, A, B):
        if A.val == B.val:
            return self.recur(A, B)
        else:
            return -1



# Solution approach:
# Recursively check if the left and right child is identical.
# If yes, then recursively check for the left child and right child.
# Else, Check if the left child of A is identical to Right child of B and Right child of A with Left child of B. 
# In this case, increment the count by one and swap the left and right pointer, check recursively.
# If both the above case fails for any node, means it is impossible to make the tree identical. Return -1.
# You need to return the count of swaps required.

#for approach used earlier by me (wrong one)-->
# You cannot keep the global count variable as it will count many irrelavant iterations also suppose
#  you tried to swap but you din't get the result as required that will be also counted in this. 
#  The implementation is not correct I have sent you proper implementation on slack
```
