```python
# Length of longest consecutive ones
# Given a binary string A. It is allowed to do at most one swap between any 0 and 1. Find and return the length of the 
# longest consecutive 1’s that can be achieved.

# Input Format
# The only argument given is string A.
# Output Format
# Return the length of the longest consecutive 1’s that can be achieved.
# Constraints
# 1 <= length of string <= 1000000
# A contains only characters 0 and 1.
# For Example
# Input 1:
#     A = "111000"
# Output 1:
#     3
# Input 2:
#     A = "111011101"
# Output 2:
#     7
```

```python
class Solution:
    # @param A : string
    # @return an integer
    def solve(self, A):
        length=len(A)
        pre=[None]* length
        suff=[None] * length
        res=0
        a=[]
        count=0
        count1=0
        for i in range(length):
            if i==0 and int(A[i])==0:
                pre[i]=0
            elif i==0 and int(A[i])!=0:
                pre[i]=1
            elif i!=0 and int(A[i])!=0:
                pre[i]=pre[i-1] + int(A[i])
            else:
                pre[i]=0

        for i in range(length-1,-1,-1):
            if i==length-1 and int(A[i])==0:
                suff[i]=0
            elif i==length-1 and int(A[i])!=0:
                suff[i]=1
            elif  int(A[i])!=0:
                suff[i]=suff[i+1] + int(A[i])
            else:
                suff[i]=0

        for i in range(length):
            if int(A[i])==1:
                count1+=1
               
        for i in range(length):
            if int(A[i])==0 and i!=0 and i!=length-1:
                if pre[i-1]+suff[i+1]<count1:
                    res=max(suff[i+1]+pre[i-1]+1,res)
                else:
                    res=max(suff[i+1]+pre[i-1],res)

            elif int(A[i])==0 and i==0 :

                if suff[i+1]<count1:
                    res=max(suff[i+1]+1,res)
                else:
                    res=max(suff[i+1],res)

            elif int(A[i])==0 and i==length-1 :
                if pre[i-1]<count1:
                    res=max(pre[i-1]+1,res)
                else:
                    res=max(pre[i-1],res)
        
        if res==0:
            res=pre[length-1]


        return res
```

```python
Christmas Trees
Problem Description
You are given an array A consisting of heights of Christmas trees and an array B of the same size consisting of the cost 
of each of the trees (Bi is the cost of tree Ai, where 1 ≤ i ≤ size(A)), and you are supposed to choose 3 trees (let's say, indices p, q, and r), such that Ap < Aq < Ar, where p < q < r.
The cost of these trees is Bp + Bq + Br.
You are to choose 3 trees such that their total cost is minimum. Return that cost.
If it is not possible to choose 3 such trees return -1.

Problem Constraints
1 <= A[i], B[i] <= 109
3 <= size(A) = size(B) <= 3000
Input Format
First argument is an integer array A.
Second argument is an integer array B.

Output Format
Return an integer denoting the minimum cost of choosing 3 trees whose heights are strictly in increasing order, if not possible, -1.

Example Input
Input 1:
 A = [1, 3, 5]
 B = [1, 2, 3]
Input 2:
 A = [1, 6, 4, 2, 6, 9]
 B = [2, 5, 7, 3, 2, 7]

Example Output
Output 1:
 6 
Output 2:
 7 

Example Explanation
Explanation 1:
 We can choose the trees with indices 1, 2 and 3, and the cost is 1 + 2 + 3 = 6. 
```

```python
class Solution:
    # @param A : list of integers
    # @param B : list of integers
    # @return an integer
    def solve(self, A, B):
        length=len(A)
        sum=10**20
        for q in range(length):
            left=0
            right=0
            sum1=10**10
            sum2=10**10
            for p in range(q):
                if A[p]<A[q]:
                    left+=1
                    sum1=min(sum1,B[p])

            for r in range(q+1,length):
                if A[q]<A[r]:
                    right+=1
                    sum2=min(sum2,B[r])
            if left>0 and right>0 :
                sum=min(sum1+B[q]+sum2,sum)
        
        if sum==10**20:
            sum=-1
        return sum
```

```python
# Spiral Order Matrix II
# Problem Description
# Given an integer A, generate a square matrix filled with elements from 1 to A2 in spiral order.

# Problem Constraints
# 1 <= A <= 1000

# Input Format
# First and only argument is integer A

# Output Format
# Return a 2-D matrix which consists of the elements in spiral order.


# Example Input
# Input 1:
# 1
# Input 2:
# 2

# Example Output
# Output 1:
# [ [1] ]
# Output 2:
# [ [1, 2], [4, 3] ]

# Example Explanation
# Explanation 1:
# Only 1 is to be arranged.
# Explanation 2:
# 1 --> 2
#       |
#       |
# 4<--- 3
```

```python
class Solution:
	# @param A : integer
	# @return a list of list of integers
	def generateMatrix(self, A):
        t=1
        count=A**2
        rows=0 #rowstart
        rowe=A-1 #rowend
        cols=0 #columnstart
        cole=A-1 #columnend
        count=0
        res=[[1 for i in range(A)]for j in range(A)]
        while rows<=rowe and cols<=cole:
            for i in range(cols,cole+1):
                count+=1
                res[rows][i]=count
            rows+=1

            for j in range(rows,rowe+1):
                count+=1
                res[j][cole]=count
            cole-=1

            for k in range(cole,cols-1,-1):
                count+=1
                res[rowe][k]=count
            rowe-=1

            for p in range(rowe,rows-1,-1):
                count+=1
                res[p][cols]=count
            cols+=1

             
        return res
```

```python
# Maximum positivity
# Problem Description
# Given an array of integers A, of size N.
# Return the maximum size subarray of A having only non-negative elements. If there are more than one such subarray, 
# return the one having the smallest starting index in A.
# NOTE: It is guaranteed that an answer always exists.

# Problem Constraints
# 1 <= N <= 105
# -109 <= A[i] <= 109

# Input Format
# The first and only argument given is the integer array A.
# Output Format
# Return maximum size subarray of A having only non-negative elements. If there are more than one such subarrays,
# return the one having earliest starting index in A.

# Example Input
# Input 1:
#  A = [5, 6, -1, 7, 8]
# Input 2:
#  A = [1, 2, 3, 4, 5, 6]

# Example Output
# Output 1:
#  [5, 6]
# Output 2:
#  [1, 2, 3, 4, 5, 6]

# Example Explanation
# Explanation 1:
#  There are two subarrays of size 2 having only non-negative elements.
#  1. [5, 6]  starting point  = 0
#  2. [7, 8]  starting point  = 3
#  As starting point of 1 is smaller, return [5, 6]
# Explanation 2:
#  There is only one subarray of size 6 having only non-negative elements:
#  [1, 2, 3, 4, 5, 6]
```

```python
class Solution:
    # @param A : list of integers
    # @return a list of integers
    def solve(self, A):
        ans=0
        count=0
        start=0
        start1=0
        end1=0
        n=len(A)
        res2=[]
        for i in range(n):
            if A[i]<0:
                start1=i+1
                count=0
            elif A[i]>=0:
                count+=1

            if count>ans:
                ans=count
                start=start1
        for i in range(start,start+ans):
            res2.append(A[i])

        return res2
            
```
